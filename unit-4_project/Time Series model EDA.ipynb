```python
import pandas as pd
```

```python
df = pd.read_csv("./data/train.csv",parse_dates=[2,3])
```

```python
df.info()
```

```python
df = df.assign(
         drp_rnd_long = round(df.dropoff_longitude,2),
         drp_rnd_lat = round(df.dropoff_latitude,2) 
         )
```

```python
df = pd.concat(
    [df,
     df.apply(lambda r: str(r['drp_rnd_long']) + '_' + str(r['drp_rnd_lat']), axis=1 ).rename('drp_loc')], 
    axis=1)
```

```python
df['drp_yyyymmdd'] = df.apply(lambda r: r['dropoff_datetime'].strftime("%Y%m%d"), axis=1)
```

```python
from datetime import datetime
def dtm_round_to_mins(dtm, p_i_min):       
    mi = str((dtm.minute // p_i_min) * p_i_min).zfill(2)
    return (datetime.strptime(dtm.strftime('%Y%m%d%H')+mi,'%Y%m%d%H%M'))
```

```python
df.loc[0]['dropoff_datetime']
```

```python
dtm_round_to_mins(df.loc[0]['dropoff_datetime'],15)
```

```python
df['drp_dtm'] = df.apply(lambda r: dtm_round_to_mins(r['dropoff_datetime'],15), axis=1)
```

```python
df.info()
```

```python
d = df.groupby(['drp_dtm','drp_yyyymmdd','drp_hrmi','drp_loc'], as_index=False)
```

```python
d.count()
```

```python
dfp = d.dropoff_datetime.count().rename(columns={'dropoff_datetime':'trips_count'})
```
